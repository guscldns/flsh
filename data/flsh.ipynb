### 데이터프레임으로 변환

```python
df = pd.DataFrame(label['images'])
df
```

```python
an = pd.DataFrame(label['annotations'])
an
```

```python
cate = pd.DataFrame(label['categories'])
cate
```

### 데이터 프레임 합치기

```python
# 물고기 있는 데이터와 라벨 값 데이터 프레임 합치기
tr = an.merge(cate, left_on='category_id', right_on='id', how='left')
tr
```

```python
df_1 = df.copy()
label_df = df_1.merge(tr, left_on='id', right_on='image_id', how='left')
label_df
```

```python
# 데이터 정보 확인
label_df.info()
```

```python
# # # NaN값 있을때는 nan값 채우거나 삭제해야 int 변환가능(nan = 무한대)
# # NaN값 있을 때
# label_df['image_id'] = label_df['image_id'].fillna(-1).astype(int)
# label_df['image_id'] = label_df['image_id'].astype(int)
# label_df['category_id'] = label_df['category_id'].fillna(-1).astype(int)
# label_df['category_id'] = label_df['category_id'].astype(int)
# label_df['iscrowd'] = label_df['iscrowd'].fillna(-1).astype(int)
# label_df['iscrowd'] = label_df['iscrowd'].astype(int)
# label_df

# Nan 값 삭제
label_df = label_df.dropna()
```

```python
# NaN값 없을 때
label_df['image_id'] = label_df['image_id'].astype(int)
label_df['category_id'] = label_df['category_id'].astype(int)
label_df['iscrowd'] = label_df['iscrowd'].astype(int)
label_df
```

```python
print(label_df['iscrowd'].value_counts(),
      label_df['segmentation'].value_counts(),
      label_df['coco_url'].value_counts(),
      label_df['flickr_url'].value_counts(),
      label_df['date_captured'].value_counts(),
      label_df['license'].value_counts(),
      label_df['supercategory'].value_counts())
```
